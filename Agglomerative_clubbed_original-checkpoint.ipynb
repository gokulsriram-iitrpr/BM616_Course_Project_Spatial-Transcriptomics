```python
import numpy as np
import pandas as pd
import scanpy as sc
import h5py
from scipy.sparse import csc_matrix
import pyarrow 
import umap


def load_dataset(base_path):
    cells = pd.read_parquet(f"{base_path}/cells.parquet")
    transcripts = pd.read_parquet(f"{base_path}/transcripts.parquet")
    metrics = pd.read_csv(f"{base_path}/metrics_summary.csv")
    meta = pd.read_csv(f"{base_path}/cells.csv.gz")

    with h5py.File(f"{base_path}/cell_feature_matrix.h5", "r") as h5:
        m = h5["matrix"]
        counts = csc_matrix(
            (m["data"][:], m["indices"][:], m["indptr"][:]),
            shape=m["shape"][:]
        )
        genes = h5["matrix"]["features"]["name"][:].astype(str)
        barcodes = h5["matrix"]["barcodes"][:].astype(str)

    adata = sc.AnnData(counts.T)
    adata.var_names = genes
    adata.obs_names = barcodes

    meta = meta.set_index("cell_id")
    adata.obs = adata.obs.join(meta)

    return adata
```

```python
adata_16week = load_dataset('C:/Users/Gokul Satya Sriram/OneDrive/Desktop/IIT ROPAR/BM616 Machine Learning/Project/Mouse adipose/16Week')
adata_4week  = load_dataset('C:/Users/Gokul Satya Sriram/OneDrive/Desktop/IIT ROPAR/BM616 Machine Learning/Project/Mouse adipose/4Week')
```

```python
adata_combined = sc.concat(
    [adata_4week, adata_16week],
    label="timepoint",
    keys=["4week", "16week"]
)
```

```python
# preprocessing

def hierarchical_preprocess(adata):

    adata.var["mt"] = adata.var_names.str.startswith(("mt-", "MT-"))
    sc.pp.calculate_qc_metrics(adata, qc_vars=["mt"], inplace=True)
    sc.pp.filter_cells(adata, min_counts=50)
    sc.pp.filter_cells(adata, min_genes=40)

    # Normalization 
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)

    # PCA (original)
    sc.pp.pca(adata, n_comps=15)

    # Extract PCA matrix for clustering
    X = adata.obsm["X_pca"]

    return adata, X
```

```python
adata_4week, X_4 = hierarchical_preprocess(adata_4week)
adata_16week, X_16 = hierarchical_preprocess(adata_16week)
adata_combined, X_combined = hierarchical_preprocess(adata_combined)
```

```python
# agglomerative clustering

from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

Z_4 = linkage(X_4, method='ward')
Z_16 = linkage(X_16, method='ward')
Z = linkage(X_combined, method='ward')
```

```python
#DENDROGRAM

plt.figure(figsize=(10, 5))
dendrogram(Z_4, truncate_mode='lastp', p=30)
plt.title("4 Week Dendrogram")
plt.show()

plt.figure(figsize=(10, 5))
dendrogram(Z_16, truncate_mode='lastp', p=30)
plt.title("16 Week Dendrogram")
plt.show()

plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=30)
plt.title("Combined Dendrogram")
plt.show()
```

```python
# Agglomerative Clustering

dist_4 = Z_4[:, 2]
threshold_4 = 0.52 * np.max(dist_4)
clusters_4 = fcluster(Z_4, t=threshold_4, criterion='distance')
adata_4week.obs["clusters"] = clusters_4.astype(str)

dist_16 = Z_16[:, 2]
threshold_16 = 0.6 * np.max(dist_16)
clusters_16 = fcluster(Z_16, t=threshold_16, criterion='distance')
adata_16week.obs["clusters"] = clusters_16.astype(str)

distances = Z[:, 2]
threshold = 0.435 * np.max(distances)
clusters = fcluster(Z, t=threshold, criterion='distance')
adata_combined.obs["clusters"] = clusters.astype(str)
```

```python
#UMAP

sc.pp.neighbors(adata_4week)
sc.tl.umap(adata_4week)
sc.pl.umap(adata_4week, color=["clusters"], title="4 Week Clusters")

sc.pp.neighbors(adata_16week)
sc.tl.umap(adata_16week)
sc.pl.umap(adata_16week, color=["clusters"], title="16 Week Clusters")

sc.pp.neighbors(adata_combined)
sc.tl.umap(adata_combined)
sc.pl.umap(
    adata_combined,
    color=["clusters", "timepoint"],
    title=["Agglomerative Clusters", "Timepoint"]
)
```

```python
# Clustering Scores:

from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

labels_4 = adata_4week.obs["clusters"].astype(int)
X = X_4

sil_score = silhouette_score(X, labels_4)
dbi_score = davies_bouldin_score(X, labels_4)
chi_score = calinski_harabasz_score(X, labels_4)

print("4 week Silhouette Score:", sil_score)
print("4 week Davies-Bouldin Index:", dbi_score)
print("4 week Calinski-Harabasz Index:", chi_score)

print("--------------------------------------------")

labels_16 = adata_16week.obs["clusters"].astype(int)
X = X_16

sil_score = silhouette_score(X, labels_16)
dbi_score = davies_bouldin_score(X, labels_16)
chi_score = calinski_harabasz_score(X, labels_16)

print("16 week Silhouette Score:", sil_score)
print("16 week Davies-Bouldin Index:", dbi_score)
print("16 week Calinski-Harabasz Index:", chi_score)

print("--------------------------------------------")

labels_combined = adata_combined.obs["clusters"].astype(int)
X = X_combined

sil_score = silhouette_score(X, labels_combined)
dbi_score = davies_bouldin_score(X, labels_combined)
chi_score = calinski_harabasz_score(X, labels_combined)

print("Combined Silhouette Score:", sil_score)
print("Combined Davies-Bouldin Index:", dbi_score)
print("Combined Calinski-Harabasz Index:", chi_score)

print("--------------------------------------------")
print("--------------------------------------------")

print("4 Week - Number of clusters:", len(np.unique(clusters_4)))
print("16 Week - Number of clusters:", len(np.unique(clusters_16)))
print("Combined - Number of clusters:", len(np.unique(clusters)))
```
